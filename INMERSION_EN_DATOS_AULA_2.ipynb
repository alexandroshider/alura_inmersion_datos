Existen otras librerias para visualizar los datos de pandas. Ejemplo de ello son las librerias de matplotlib y seaborn. Esta ultima posee más detalle y se pueden hacer más bonitas.

```python
#importamos las librerías para graficar
import matplotlib.pyplot as plt
import seaborn as sns

#Definimos la superficie de la visualización 10 para horizontar y 8 vertical
plt.figure(figsize=(10,6))
#hacer historias usando seaborn
grafica = sns.histplot(data=inmuebles, x='Precio_Millon', 
                       kde=True, #kde nos permite ver la curva de densidad en la gráfica 
                       hue='Tipo') #Se puede separar en los diferentes tipos de inmuebles  
grafica.set_title("Distribución de los precios de inmuebles en Bogotá") #titulo de grafica
plt.xlim(50,1000) #Limitamos los valores minimos y máximos de los inmuebles 
                  #para evitar tener los valores más altos y que hay poca densidad
                  #en esos valores.
#plt.ylim(0,20)#Para ver distribución de inmuebles
#Guardar la imagen del grafico                  
plt.savefig('/content/drive/MyDrive/valores_inmuebles.png',format='png')
plt.show()
```

# Desafio 2: Precio del m**2 por barrio





```python
#Hacer la columna directamente en la tabla de inmuebles es un error 
#común. Ya que estaríamos promediando sobre una división. 
#inmuebles['Precio/m**2'] = inmuebles.Precio_Millon/inmuebles.Area
#inmuebles
```

```python
barrios = inmuebles.groupby('Barrio').sum()
#barrios
barrios['Precio_m^2_Millon'] = barrios.Precio_Millon/barrios.Area
barrios
```

```python
import matplotlib.pyplot as plt
#barrios = inmuebles.groupby(['Barrio']).mean() #Agrupar inmuebles por barrios
#Grafica
barrios_sort = barrios.sort_values(by="Precio_m^2_Millon")
plt.figure(figsize=(15,9))
fig,ax=plt.subplots()
x = barrios.index
y = barrios["Precio_m^2_Millon"]
plt.barh(x,y)
plt.ylabel('Barrio')
plt.xlabel('Precio m^2 en Millones')
ax.set_title('How fast do you want to go today?')
#plt.title(3)
plt.show()
```

```python
#barrios = inmuebles.groupby(['Barrio']).mean() #Agrupar inmuebles por barrios
#Grafica
barrios_sorted = barrios.sort_values(by="Precio_m^2_Millon",ascending=False).head(30)
plt.figure(figsize=(15,9))
#x = barrios.index
#y = barrios["Precio_m^2_Millon"]

grafica = sns.barplot(data=barrios_sorted, y=barrios_sorted.index, x='Precio_m^2_Millon',orient='h',
            palette="Spectral",label="Hola").set_title('''Precio promedio del \
m^2 por barrio''', fontdict={'size': 30})

plt.ylabel('Barrio',fontsize=20)
plt.xlabel('Precio del m^2 en Millones',fontsize=20)
plt.savefig('/content/drive/MyDrive/valores_inmuebles.png',format='png')
#plt.title(3)
plt.show()
```
